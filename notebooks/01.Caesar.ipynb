```python
def caesar_cipher(text, shift, mode='encrypt'):
    result = ""

    # Αν κάνουμε αποκρυπτογράφηση, απλά αντιστρέφουμε το πρόσημο του shift
    if mode == 'decrypt':
        shift = -shift

    for char in text:
        # Έλεγχος αν είναι γράμμα (ώστε να μην πειράξουμε κενά, τελείες, κλπ)
        if char.isalpha():

            # Καθορισμός της βάσης (offset)
            # Αν είναι κεφαλαίο -> 65 ('A'), αν είναι μικρό -> 97 ('a')
            start = ord('A') if char.isupper() else ord('a')

            # Βήμα 1: Μετατροπή ASCII σε 0-25
            original_position = ord(char) - start

            # Βήμα 2: Shift και Modulo (Ο πυρήνας του αλγορίθμου)
            new_position = (original_position + shift) % 26

            # Βήμα 3: Μετατροπή 0-25 πίσω σε ASCII
            new_char = chr(new_position + start)

            result += new_char
        else:
            # Αν δεν είναι γράμμα, το αφήνουμε όπως είναι
            result += char

    return result
```

```python
# --- Main Program ---
message = input("Δώσε το μήνυμα: ")
key = int(input("Δώσε το κλειδί μετατόπισης (ακέραιος): "))

# Κρυπτογράφηση
encrypted = caesar_cipher(message, key, mode='encrypt')
print(f"\nEncrypted: {encrypted}")

# Αποκρυπτογράφηση (Επαλήθευση)
decrypted = caesar_cipher(encrypted, key, mode='decrypt')
print(f"Decrypted: {decrypted}")
```
