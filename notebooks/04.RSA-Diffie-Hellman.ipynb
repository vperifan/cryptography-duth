```python
def mod_exp(a, e, n):
    result = 1
    a = a % n

    while e > 0:
        # Αν ο εκθέτης είναι περιττός, πολλαπλασιάζουμε το αποτέλεσμα με τη βάση
        if e % 2 == 1:
            result = (result * a) % n

        # Υψώνουμε τη βάση στο τετράγωνο και παίρνουμε το modulo
        a = (a * a) % n

        # Υποδιπλασιάζουμε τον εκθέτη (ακέραια διαίρεση)
        e //= 2

    return result

# --- Απαντήσεις Άσκησης 1 ---
print("--- Πρόβλημα 1: Modular Exponentiation ---")
print(f"3^200 mod 13 = {mod_exp(3, 200, 13)}")       # Αναμενόμενο: 9
print(f"5^117 mod 19 = {mod_exp(5, 117, 19)}")       # Αναμενόμενο: 1
print(f"7^560 mod 561 = {mod_exp(7, 560, 561)}")     # Αναμενόμενο: 1
print("-" * 40)
```

```python
def extended_gcd(a, b):
    # Βασική περίπτωση: αν το a είναι 0, ο ΜΚΔ είναι το b
    if a == 0:
        return b, 0, 1

    gcd, x1, y1 = extended_gcd(b % a, a)

    # Ανανέωση των x και y με βάση τα αποτελέσματα της αναδρομής
    x = y1 - (b // a) * x1
    y = x1

    return gcd, x, y

def mod_inverse(a, n):
    gcd, x, y = extended_gcd(a, n)

    if gcd != 1:
        # Αν ο ΜΚΔ δεν είναι 1, δεν υπάρχει αντίστροφο
        return None
    else:
        # Επιστρέφουμε το x % n για να σιγουρευτούμε ότι είναι θετικό
        return x % n

# --- Απαντήσεις Άσκησης 2 ---
print("--- Πρόβλημα 2: Modulo Inverse ---")
print(f"3^-1 mod 11 = {mod_inverse(3, 11)}")         # Αναμενόμενο: 4 (γιατί 3*4 = 12 = 1 mod 11)
print(f"10^-1 mod 17 = {mod_inverse(10, 17)}")       # Αναμενόμενο: 12 (γιατί 10*12 = 120 = 1 mod 17)
print(f"17^-1 mod 3120 = {mod_inverse(17, 3120)}")   # Αναμενόμενο: 2753
print("-" * 40)
```

# RSA
- Έστω p=3,q=11. Άρα n=33 και ϕ(n)=(3-1)(11-1)=20.
- Έστω δημόσιο κλειδί e=3.
- Βρίσκουμε το ιδιωτικό κλειδί d: 3⋅d≡1(mod20)⇒d=7.
- Μήνυμα m=4.
- Κρυπτογράφηση: c=43(mod33)=64(mod33)=31.
- Αποκρυπτογράφηση: m=317(mod33).

```python
# Βήμα 1: Επιλογή Πρώτων (εδώ τους δίνουμε έτοιμους)
p = 3
q = 11
print(f"1. Επιλέξαμε p={p}, q={q}")

# Βήμα 2: Υπολογισμός n και φ(n)
n = p * q
phi_n = (p - 1) * (q - 1)
print(f"2. Υπολογίσαμε n = {n} και φ(n) = {phi_n}")

# Βήμα 3: Επιλογή Δημόσιου Κλειδιού e
e = 3
print(f"3. Επιλέξαμε δημόσιο κλειδί e = {e}")

# Βήμα 4: Υπολογισμός Ιδιωτικού Κλειδιού d
# Εδώ καλούμε τη συνάρτηση από το Πρόβλημα 2 του προηγούμενου εργαστηρίου!
d = mod_inverse(e, phi_n)
print(f"4. Υπολογίσαμε το ιδιωτικό κλειδί d = {e}^-1 mod {phi_n} = {d}")

print("\n--- Κλειδιά ---")
print(f"Δημόσιο Κλειδί (Public Key): (e={e}, n={n})")
print(f"Ιδιωτικό Κλειδί (Private Key): (d={d}, n={n})")
print("----------------\n")

# Βήμα 5: Κρυπτογράφηση
m = 4  # Το αρχικό μας μήνυμα
print(f"Το αρχικό μήνυμα είναι: m = {m}")

# Κρυπτογράφηση: c = m^e mod n (Καλούμε τη συνάρτηση από το Πρόβλημα 1 του προηγούμενου εργαστηρίου)
c = mod_exp(m, e, n)
print(f"Κρυπτογράφηση: Το κρυπτογραφημένο μήνυμα (ciphertext) είναι c = {c}")

# Βήμα 6: Αποκρυπτογράφηση
# Αποκρυπτογράφηση: m_decrypted = c^d mod n
m_decrypted = mod_exp(c, d, n)
print(f"Αποκρυπτογράφηση: Το αποκρυπτογραφημένο μήνυμα είναι m = {m_decrypted}")

# Τελικός Έλεγχος
if m == m_decrypted:
    print("\nΕπιτυχία! Τα μαθηματικά του RSA λειτουργούν τέλεια.")
else:
    print("\nΚάτι πήγε στραβά...")
```

Diffie-Hellman

```python
print("=== Ανταλλαγή Κλειδιών Diffie-Hellman ===")

# 1. Δημόσιες Παράμετροι (Public)
p = 23
g = 5
print(f"Δημόσια δεδομένα: p={p}, g={g}\n")

# 2. Η πλευρά της Αλίκης
a = 4  # Μυστικό κλειδί Αλίκης
A = mod_exp(g, a, p)  # A = g^a mod p
print(f"Η Αλίκη διάλεξε a={a}.")
print(f"Η Αλίκη υπολογίζει το δημόσιο A = {g}^{a} mod {p} = {A}")
print("Η Αλίκη στέλνει το Α στον Μπομπ.\n")

# 3. Η πλευρά του Μπομπ
b = 3  # Μυστικό κλειδί Μπομπ
B = mod_exp(g, b, p)  # B = g^b mod p
print(f"Ο Μπομπ διάλεξε b={b}.")
print(f"Ο Μπομπ υπολογίζει το δημόσιο B = {g}^{b} mod {p} = {B}")
print("Ο Μπομπ στέλνει το B στην Αλίκη.\n")

# 4. Υπολογισμός Κοινού Μυστικού (Shared Secret)
# Η Αλίκη χρησιμοποιεί το B του Μπομπ και το δικό της a
secret_Alice = mod_exp(B, a, p)  # S = B^a mod p

# Ο Μπομπ χρησιμοποιεί το A της Αλίκης και το δικό του b
secret_Bob = mod_exp(A, b, p)    # S = A^b mod p

print("--- Υπολογισμός Μυστικού Κλειδιού ---")
print(f"Η Αλίκη υπολογίζει: S_A = {B}^{a} mod {p} = {secret_Alice}")
print(f"Ο Μπομπ υπολογίζει: S_B = {A}^{b} mod {p} = {secret_Bob}")

if secret_Alice == secret_Bob:
    print(f"\nΕπιτυχία! Και οι δύο συμφώνησαν στο μυστικό κλειδί: {secret_Alice}")
    print("Η Eve που βλέπει τα p=23, g=5, A=4, B=10 ΔΕΝ μπορεί να βρει εύκολα το 18!")
else:
    print("\nΣφάλμα στους υπολογισμούς.")
```

```python
from cryptography.hazmat.primitives.asymmetric import dh
from cryptography.hazmat.primitives.kdf.hkdf import HKDF
from cryptography.hazmat.primitives import hashes
from cryptography.hazmat.primitives import serialization

print("=== Σενάριο 1: Ανταλλαγή Κλειδιών Diffie-Hellman ===\n")

# Βήμα 1: Δημιουργία των Δημόσιων Παραμέτρων (p, g)
# Σε πραγματικές συνθήκες αυτό αργεί, οπότε διαλέγουμε ένα μικρό key size για το εργαστήριο (π.χ. 512, κανονικά > 2048)
# Συνήθως το g είναι 2 ή 5.
print("1. Παραγωγή ασφαλών παραμέτρων DH (p, g)... Παρακαλώ περιμένετε.")
parameters = dh.generate_parameters(generator=2, key_size=512)

params = parameters.parameter_numbers()
p = params.p
g = params.g

print("p (prime modulus):", p)
print(p)

print("\ng (generator):", g)

# Βήμα 2: Η Αλίκη δημιουργεί τα κλειδιά της
alice_private_key = parameters.generate_private_key()
alice_public_key = alice_private_key.public_key()
print("2. Η Αλίκη παρήγαγε το Ιδιωτικό και το Δημόσιο κλειδί της.")

# Βήμα 3: Ο Μπομπ δημιουργεί τα κλειδιά του (με τις ίδιες παραμέτρους)
bob_private_key = parameters.generate_private_key()
bob_public_key = bob_private_key.public_key()
print("3. Ο Μπομπ παρήγαγε το Ιδιωτικό και το Δημόσιο κλειδί του.")

alice_public_bytes = alice_public_key.public_bytes(
    encoding=serialization.Encoding.PEM,
    format=serialization.PublicFormat.SubjectPublicKeyInfo
)

bob_public_bytes = bob_public_key.public_bytes(
    encoding=serialization.Encoding.PEM,
    format=serialization.PublicFormat.SubjectPublicKeyInfo
)

print("\nAlice Public Key:")
print(alice_public_bytes.decode())

print("Bob Public Key:")
print(bob_public_bytes.decode())

# Βήμα 4: ΑΝΤΑΛΛΑΓΗ (Μέσω του ανοιχτού WiFi)
# Η Αλίκη λαμβάνει το bob_public_key. Ο Μπομπ λαμβάνει το alice_public_key.

# Βήμα 5: Υπολογισμός Κοινού Μυστικού (Shared Secret)
shared_secret_alice = alice_private_key.exchange(bob_public_key)
shared_secret_bob = bob_private_key.exchange(alice_public_key)

print(f"\n4. Συμφωνούν τα μυστικά; {'ΝΑΙ!' if shared_secret_alice == shared_secret_bob else 'ΟΧΙ!'}")

# ΣΗΜΑΝΤΙΚΟ: Το shared_secret δεν το χρησιμοποιούμε απευθείας ως κλειδί AES!
# Το περνάμε από μια Key Derivation Function (KDF) για να γίνει ένα τέλειο κλειδί 32 bytes (256-bit).
derived_key = HKDF(
    algorithm=hashes.SHA256(),
    length=32,
    salt=None,
    info=b'handshake data',
).derive(shared_secret_alice)

print(f"5. Τελικό Κλειδί AES (Hex): {derived_key.hex()}")
```

```python
from Crypto.PublicKey import RSA
from Crypto.Cipher import PKCS1_OAEP
from Crypto.Signature import pkcs1_15
from Crypto.Hash import SHA256

print("\n=== Σενάριο 2: Πλήρης επικοινωνία με RSA ===\n")

# --- ΠΡΟΕΤΟΙΜΑΣΙΑ ---
# 1. Δημιουργία Κλειδιών (Κανονικά αυτό γίνεται μια φορά και τα αποθηκεύουμε σε αρχεία .pem)
print("1. Δημιουργία RSA Κλειδιών (2048-bit)...")
alice_key = RSA.generate(2048)  # Ιδιωτικό & Δημόσιο της Αλίκης
bob_key = RSA.generate(2048)    # Ιδιωτικό & Δημόσιο του Μπομπ

alice_pub = alice_key.publickey()
bob_pub = bob_key.publickey()

# Εκτύπωση Δημόσιων Κλειδιών σε PEM μορφή
print("\nAlice Public Key:")
print(alice_pub.export_key().decode())

print("Bob Public Key:")
print(bob_pub.export_key().decode())

message = b"Transfer 1000 euros to account GR123456789"
print(f"Αρχικό Μήνυμα: {message.decode('utf-8')}")

# --- Η ΠΛΕΥΡΑ ΤΗΣ ΑΛΙΚΗΣ (Αποστολέας) ---
# Βήμα 1: Υπογραφή (Signature)
# Η Αλίκη παίρνει το Hash του μηνύματος και το υπογράφει με το ΔΙΚΟ ΤΗΣ ιδιωτικό κλειδί
msg_hash = SHA256.new(message)
signature = pkcs1_15.new(alice_key).sign(msg_hash)
print("\n[Αλίκη] Δημιούργησε την ψηφιακή υπογραφή.")

print("\nΨηφιακή Υπογραφή (hex):")
print(signature.hex())

# Βήμα 2: Κρυπτογράφηση (Encryption)
# Η Αλίκη κρυπτογραφεί το μήνυμα (και την υπογραφή, αν θέλει) με το ΔΗΜΟΣΙΟ κλειδί του ΜΠΟΜΠ
cipher_rsa = PKCS1_OAEP.new(bob_pub)
encrypted_message = cipher_rsa.encrypt(message)
print(f"[Αλίκη] Κρυπτογράφησε το μήνυμα. (Ciphertext: {encrypted_message[:15].hex()}...)")

# Η Αλίκη στέλνει μέσω internet το (encrypted_message, signature)


# --- Η ΠΛΕΥΡΑ ΤΟΥ ΜΠΟΜΠ (Παραλήπτης) ---
print("\n--- Το πακέτο φτάνει στον Μπομπ ---")

# Βήμα 3: Αποκρυπτογράφηση (Decryption)
# Ο Μπομπ αποκρυπτογραφεί το μήνυμα με το ΔΙΚΟ ΤΟΥ ιδιωτικό κλειδί
decipher_rsa = PKCS1_OAEP.new(bob_key)
try:
    decrypted_message = decipher_rsa.decrypt(encrypted_message)
    print(f"[Μπομπ] Αποκρυπτογράφησε το μήνυμα επιτυχώς: {decrypted_message.decode('utf-8')}")
except ValueError:
    print("[Μπομπ] Σφάλμα: Η αποκρυπτογράφηση απέτυχε!")

# Βήμα 4: Επαλήθευση Υπογραφής (Verification)
# Ο Μπομπ παίρνει το Hash του αποκρυπτογραφημένου μηνύματος
# και ελέγχει την υπογραφή χρησιμοποιώντας το ΔΗΜΟΣΙΟ κλειδί της ΑΛΙΚΗΣ
received_hash = SHA256.new(decrypted_message)
verifier = pkcs1_15.new(alice_pub)

try:
    verifier.verify(received_hash, signature)
    print("[Μπομπ] Η υπογραφή είναι ΕΓΚΥΡΗ! Το μήνυμα είναι 100% από την Αλίκη και δεν αλλοιώθηκε.")
except (ValueError, TypeError):
    print("[Μπομπ] ΚΙΝΔΥΝΟΣ! Η υπογραφή ΔΕΝ είναι έγκυρη. Το μήνυμα ίσως είναι πλαστό.")
```
