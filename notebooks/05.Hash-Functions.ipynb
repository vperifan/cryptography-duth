```python
def mini_hash(message):
    # 1. Αρχικές σταθερές
    hash_value = 0xABCD   # Αρχική τιμή (IV)
    prime = 31            # Πρώτος αριθμός για ανακάτεμα
    mod = 65536           # 2^16, για να κρατήσουμε το αποτέλεσμα στα 16-bits

    # 2. Επεξεργασία κάθε χαρακτήρα
    for char in message:
        ascii_val = ord(char)

        # Ανάμειξη με XOR
        hash_value = hash_value ^ ascii_val

        # Πολλαπλασιασμός και Modulo
        hash_value = (hash_value * prime) % mod

    # Επιστρέφουμε τη δεκαεξαδική του μορφή, αφαιρώντας το '0x' και
    # γεμίζοντας με μηδενικά μπροστά αν χρειαστεί (zfill)
    return hex(hash_value)[2:].zfill(4).upper()


# --- Εκτέλεση Πειραμάτων ---
print("=== Πειράματα MiniHash-16 ===")

hash1 = mini_hash("HELLO")
hash2 = mini_hash("HELLP")
hash3 = mini_hash("hello")
hash4 = mini_hash("A")
hash5 = mini_hash("This is a very long sentence to prove that the hash remains the same size.")

print(f"Hash('HELLO'): {hash1}")
print(f"Hash('HELLP'): {hash2}  <-- Αλλαγή ενός γράμματος")
print(f"Hash('hello'): {hash3}  <-- Πεζά αντί για κεφαλαία")
print(f"Hash('A'):     {hash4}")
print(f"Hash(Long):    {hash5}")
```

```python
print("=== Επίθεση Σύγκρουσης (Collision) ===")

target_message = "SECURE"
target_hash = mini_hash(target_message)
print(f"Ψάχνουμε σύγκρουση για το '{target_message}' (Hash: {target_hash})")

counter = 0
while True:
    test_message = "HACK" + str(counter)
    test_hash = mini_hash(test_message)

    if test_hash == target_hash and test_message != target_message:
        print(f"[!] BINGO! Βρέθηκε σύγκρουση!")
        print(f"Το μήνυμα '{test_message}' παράγει ακριβώς το ίδιο Hash: {test_hash}")
        print(f"Χρειάστηκαν {counter} προσπάθειες.\n")
        break

    counter += 1
```

```python
# 1. Καθηγητής
original_msg = "AM: 1045, Vathmos: 5"
original_hash = mini_hash(original_msg)
print(f"[Καθηγητής] Στέλνει μήνυμα: '{original_msg}'")
print(f"[Καθηγητής] Στέλνει Hash:   {original_hash}\n")

# 2. Hacker
intercepted_msg = "AM: 1045, Vathmos: 10"
print(f"[Hacker] Αλλάζει το μήνυμα σε: '{intercepted_msg}'")
print(f"[Hacker] Προωθεί το αρχικό Hash: {original_hash}\n")

# 3. Γραμματεία
print("[Γραμματεία] Παραλαβή δεδομένων και έλεγχος...")
received_hash = mini_hash(intercepted_msg)

print(f"Υπολογισμένο Hash: {received_hash}")
print(f"Αναμενόμενο Hash:  {original_hash}")

if received_hash == original_hash:
    print("[Γραμματεία] Επιτυχία! Η βαθμολογία καταχωρήθηκε.")
else:
    print("[Γραμματεία] Το μήνυμα έχει παραποιηθεί. Η αλλαγή ακυρώνεται!")
```

```python
import hashlib

print("=== SHA-256 ===\n")

# --- Βασική Χρήση της hashlib ---
message = "HELLO"

# Δημιουργούμε το αντικείμενο κατακερματισμού
hash_object = hashlib.sha256()

# Τα Hash functions δέχονται μόνο bytes, οπότε κωδικοποιούμε το string
byte_message = message.encode('utf-8')

# Εισάγουμε τα bytes στο αντικείμενο
hash_object.update(byte_message)

# Παίρνουμε το αποτέλεσμα σε δεκαεξαδική μορφή
hash_hex = hash_object.hexdigest()

print("1. Σταθερό Μέγεθος (Fixed Length):")
print(f"SHA-256('HELLO'): {hash_hex}")
print(f"Μήκος: {len(hash_hex)} δεκαεξαδικά ψηφία (256 bits)\n")


# --- Το Φαινόμενο της Χιονοστιβάδας (Avalanche Effect) ---
# Αλλάζουμε μόνο ΕΝΑ γράμμα
message2 = "HELLP"
hash2_hex = hashlib.sha256(message2.encode('utf-8')).hexdigest()

print("2. Φαινόμενο Χιονοστιβάδας (Αλλάζουμε το O σε P):")
print(f"SHA-256('HELLP'): {hash2_hex}")

print(f"HELLO: {hash_hex}")
print(f"HELLP: {hash2_hex}")
print("Αλλάζοντας 1 bit στην είσοδο, αλλάζει σχεδόν το 50% των bits στην έξοδο!")
```
