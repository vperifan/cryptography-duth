```python
def mod_exp(a, e, n):
    result = 1
    a = a % n

    while e > 0:
        # Αν ο εκθέτης είναι περιττός, πολλαπλασιάζουμε το αποτέλεσμα με τη βάση
        if e % 2 == 1:
            result = (result * a) % n

        # Υψώνουμε τη βάση στο τετράγωνο και παίρνουμε το modulo
        a = (a * a) % n

        # Υποδιπλασιάζουμε τον εκθέτη (ακέραια διαίρεση)
        e //= 2

    return result

# --- Απαντήσεις Άσκησης 1 ---
print("--- Πρόβλημα 1: Modular Exponentiation ---")
print(f"3^200 mod 13 = {mod_exp(3, 200, 13)}")       # Αναμενόμενο: 9
print(f"5^117 mod 19 = {mod_exp(5, 117, 19)}")       # Αναμενόμενο: 1
print(f"7^560 mod 561 = {mod_exp(7, 560, 561)}")     # Αναμενόμενο: 1
print("-" * 40)
```

```python
def extended_gcd(a, b):
    # Βασική περίπτωση: αν το a είναι 0, ο ΜΚΔ είναι το b
    if a == 0:
        return b, 0, 1

    gcd, x1, y1 = extended_gcd(b % a, a)

    # Ανανέωση των x και y με βάση τα αποτελέσματα της αναδρομής
    x = y1 - (b // a) * x1
    y = x1

    return gcd, x, y

def mod_inverse(a, n):
    gcd, x, y = extended_gcd(a, n)

    if gcd != 1:
        # Αν ο ΜΚΔ δεν είναι 1, δεν υπάρχει αντίστροφο
        return None
    else:
        # Επιστρέφουμε το x % n για να σιγουρευτούμε ότι είναι θετικό
        return x % n

# --- Απαντήσεις Άσκησης 2 ---
print("--- Πρόβλημα 2: Modulo Inverse ---")
print(f"3^-1 mod 11 = {mod_inverse(3, 11)}")         # Αναμενόμενο: 4 (γιατί 3*4 = 12 = 1 mod 11)
print(f"10^-1 mod 17 = {mod_inverse(10, 17)}")       # Αναμενόμενο: 12 (γιατί 10*12 = 120 = 1 mod 17)
print(f"17^-1 mod 3120 = {mod_inverse(17, 3120)}")   # Αναμενόμενο: 2753
print("-" * 40)
```

```python
def fermat_primality_test(p, a=2):
    # Εξαιρέσεις για πολύ μικρούς αριθμούς
    if p <= 1:
        return False
    if p == 2:
        return True

    # Έλεγχος Fermat: a^(p-1) mod p == 1
    if mod_exp(a, p - 1, p) == 1:
        return True  # Πιθανότατα πρώτος
    else:
        return False # Σίγουρα σύνθετος

# --- Απαντήσεις Άσκησης 3 ---
print("--- Πρόβλημα 3: Fermat Test (base a=2) ---")
print(f"Είναι το 17 πρώτος; {fermat_primality_test(17)}")   # Αναμενόμενο: True
print(f"Είναι το 19 πρώτος; {fermat_primality_test(19)}")   # Αναμενόμενο: True
print(f"Είναι το 561 πρώτος; {fermat_primality_test(561)}") # Αναμενόμενο: True (ΠΑΓΙΔΑ: Είναι Carmichael number!)
# Στην πράξη χρειαζόμαστε πιο ισχυρά tests όπως Miller-Rabin ή/και Baillie-PSW
print("-" * 40)
```

```python
def chinese_remainder_theorem(remainders, moduli):
    total = 0

    # 1. Βρίσκουμε το γινόμενο όλων των moduli (N)
    N = 1
    for m in moduli:
        N *= m

    # 2. Εφαρμογή του τύπου CRT
    for a_i, m_i in zip(remainders, moduli):
        # Υπολογίζουμε το N_i (το γινόμενο όλων εκτός του τρέχοντος m_i)
        N_i = N // m_i

        # Βρίσκουμε το αντίστροφο του N_i modulo m_i
        y_i = mod_inverse(N_i, m_i)

        # Προσθέτουμε στο συνολικό άθροισμα
        total += a_i * N_i * y_i

    # 3. Το αποτέλεσμα είναι το total modulo N
    return total % N

# --- Απαντήσεις Άσκησης 4 ---
print("--- Πρόβλημα 4: Chinese Remainder Theorem ---")
remainders = [2, 3, 1, 3]
moduli = [3, 5, 4, 7]

x = chinese_remainder_theorem(remainders, moduli)
print(f"Η λύση του συστήματος είναι: x = {x}") # Αναμενόμενο: 53
print("Επαλήθευση:")
print(f"53 mod 3 = {53 % 3}")
print(f"53 mod 5 = {53 % 5}")
print(f"53 mod 4 = {53 % 4}")
print(f"53 mod 7 = {53 % 7}")
print("-" * 40)
```
